# Confidence Intervals

- Confidence intervals express a range of values where you can be a certain percent (typically 95%) confident that a statistic falls between
    - Statistic might be a mean, proportion, etc
- Type of score found depends on whether the standard deviation is known
    - If so: use z-score
    - If not: use t-score

## Confidence Interval for Proportion
- The Central Limit Theorem implies a normal model for the sampling distribution of $\hat{p}$
- Sample Proportion
    - $\hat{p} = \frac{x}{n}$
    - x is number out of population
    - p is population
- Mean
    - $µ_{\hat{p}}$ = *p*
- Standard Error
    - SE uses population parameters, whereas se estimates SE by substituting sample statistics for population parameters
    - $se(\hat{p}) = σ_{\hat{p}} = \sqrt{\frac{p(1-p)}{n}}$
### Calculating CI for $\hat{p}$
- $\hat{p} - z_{\frac{α}{2}}SE \leq p \leq \hat{p} + z_{\frac{α}{2}}SE$
- Fill in se($\hat{p}$):
    - $\hat{p} - z_{\frac{α}{2}}\sqrt{\frac{p(1-p)}{n}} \leq p \leq \hat{p} + z_{\frac{α}{2}}\sqrt{\frac{p(1-p)}{n}}$
- For 0.95, the p outside this range is 0.025 on either side
    - 1 - 0.05 = 0.95, 0.05 = 0.025 / 2 
- For a 95% confidence interval:
    - **$z_{\frac{0.05}{2}} = 1.96$ STDs**
    - z = `qnorm(0.025)` 
- Checklist for Confidence Interval for *p*
    - The sample is a simple random sample (SRS)
    - Both $n\hat{p}$ and $n(1-\hat{p})$ are larger than 10

## Confidence Interval for the Mean
- Mean refers to mean of sample from population
- Similar procedure that we used for *p*
- The estimated standard error (se$\bar{X}$) is different
    - $SE(\bar{X}) = σ_{\bar{X}} = \frac{s}{\sqrt{n}}$
- Instead of a z-score, we will use a t-score
    - This is to account for the unknown standard deviation of a sample
    - t-distributions have wider percentiles to make CI more accurate
    - Using R, t = `qt(0.025, n-1)`
- Degrees of freedom (df)
    - df = n - 1 
    - Controls the shape of the distribution depending on sample size *n*
    - As *n* increases, the t-distribution more closely resembles normal distribution

### Calculating Confidence Interval for µ
- $\bar{x} - t_{\frac{α}{2},n-1}se \leq µ \leq \bar{x} + t_{\frac{α}{2},n-1}se$

Fill in se($\bar{x}$):
- $\bar{x} - t_{\frac{α}{2},n-1}\frac{s}{\sqrt{n}} \leq µ \leq \bar{x} + t_{\frac{α}{2},n-1}\frac{s}{\sqrt{n}}$


- The value of t depdens on level of confidence and n-1 df
    - Lookup t based on df in t-chart
- Checklist for CI µ
    - Sample is SRS from the population
    - The sample size is larger than 10x the absolute value of the kurtosis
        - $n = 10|{K4}|$

### Confidence Interval in R
- Use `qt(alpha/2, df)`
    - "Quantile of t"
- Notice this value is negative, to get a positive use:
    - `qt(1-alpha/2, df)`

```r
df = 125
alpha = 0.05

# Get critical value
qt(alpha/2, df)

# Convert to positive critical value
qt(1-alpha/2, df)
```

We can also change the confidence interval, for example 99%:

```r
df = 125
# Use .01 for 99% CI
alpha = 0.01

# Get critical value
qt(alpha/2, df)
```

Compare with normal score using qnorm:

```r
df = 125
alpha = 0.05

# Get critical value
qnorm(alpha/2, df)
```

#### Use t.test() Function for Quicker Result

`t.test(val1, val2[optional], conf.level = .95)`

```r

data <- read.csv("https://goo.gl/ZCVUpc")
Xbar <- mean(data$Teabags)
sd <- sd(data$Teabags)
n = length(data$Teabags)

#a) Construct a 95\% confidence interval estimate for the population mean weight of the tea bags. Interpret the interval. (10 points) *

# Statistics +- CV * SE 

# Find t of (α/2) = .05/2 = 0.025
# Find lower and upper bounds
lb = Xbar - qt (0.975, df = 49) * sd/sqrt(n)
ub = Xbar + qt(0.975, df = 49) * sd/sqrt(n)
c(lb, ub) 

# Histogram  with values
#hist(rt(1000, df = 49))
#abline(v = c(qt(0.025, df = 49), qt(0.975, df = 49)), col = 'red')

# We can also use this super convenient function that our professor
#...didn't teach us until just now!! 
# It was probably a good idea because
#...I definitely would have used this lazy method instead :) 
t.test(data$Teabags, conf.level = .95)
```

### Interpreting Confidence Intervals
- Interpretations (using CC example above)
    - Wrong:
        - "95% of all customers keep a balance of $1,520 to $2,460" 
        - This is just one sample, we can't make a statement about the population
    - Corect:
        - "The mean balance of 95% of samples of this size will fall between $1,520 and $2,460"        

Simulating 10,000 samples, we see that the mu is covered in almost every lower and upper bound
- Theoretically it should be covered in 95% of these samples
- The actual mean of means within those bounds are printed at the end
- By raising the n we can get even closer to 0.95

```r
set.seed(123)

# Sample Size
n = 10
# Number of samples, each of size n
NSIM = 10000 

rv = rnorm(n * NSIM, 10, 1)

mu = 10
sd = 1
alpha = 0.05

# Put simulated numbers into matrix
# Each row is a sample
rvm = matrix(rv, ncol = n, byrow = TRUE)

# Row means = mean of each row
ybars = rowMeans(rvm)

# Workaround to create row standard deviation 
# Apply sd to each row
sdevs = apply(rvm, 1, sd)

# Find lower and upper bounds for each sample
lower = ybars - qt(1-alpha/2, n - 1) * sdevs/sqrt(n)
upper = ybars + qt(1-alpha/2, n - 1) * sdevs/sqrt(n)

# An estimate of the true confidence level of the ordinary frequentist confidnece interval:
head(cbind(ybars, sdevs, n, lower, upper, mu), 10)
mean((lower < mu) & (mu < upper))
```

## Determining Sample Size


### Margin of Error (MoE)
- Half-width of the confidence interval
    - Confidence Interval Formula:
        - $\bar{x} \pm$ $t_{\frac{α}{2},n-1}se$
    - MoE is the right half of the equation:
        - $t_{\frac{α}{2},n-1}se$
- Three impacts on MoE
    - Level of confidence (α)
    - Variation in data (se)
    - Sample size (n)
- Formula
    - Margin of Error = $2\frac{s}{\sqrt{n}}$
        - Used for 95% confidence intervals
    - Deriving this replaces $t_{\frac{.05}{2}}, n-1$ with 2
        - Don't ask me how, staticians apparently play fast and loose with math
        - Using some dark magic algebra we arrive at the MoE formula

### Finding sample size with MoE
- $n = \frac{4σ^{2}}{(MoE)^{2}}$
- We obtain an estimate for $σ^{2}$ by using a pilot sample
    - Pilot samples are small sample of 10-30ish cases to estimate σ
- When looking for *p*, there is no need for a pilot sample
    - Use p = 0.5 which results in largest possible value for *n*
    - With this and more dark magic algebra we get the formula:
    - $n = \frac{1}{e^2}$

## Applications of Confidence Interval

### Example: Property Taxes
*A mayor is considering a tax on business that is proportional to the amount spent to lease property in his city.*

- Cost of Lease is \$478,603.48
- Standard deviation: 535,342.56
- Sample size: 223
- *How much revenue would a 1\% tax generate?*


- We need a confidence interval on the amount of leased property
    - $\bar{X}$ = $478,603.48
    - S = $535,342.56

```r
xbar = 478603.48

s = 535342.56

n = 223

# Calculate standard error
se = s/sqrt(n)

alpha = 1-.95

# Find lower bound
lower = xbar - qt(1-alpha/2, n-1) * se
lower

# Find upper bound
upper = xbar + qt(1-alpha/2, n-1) * se
upper
```

- We are now 95% confident the average cost of a lease is between \\$410,000 and \\$550,000
- Tax revenue is 1\% 
- Therefore 95% confidence interval for tax raised per lease is \\$4,100 to \\$5,500
