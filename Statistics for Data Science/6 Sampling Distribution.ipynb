# Sampling Distribution

## Overview

- Making a conclusion from a sample of a population
    - Conclusions include statistics of the sample
- Import to ensure the sample is representative of the entire population
- Population
    - The entire collection of the interested data
    - E.g. to study height of adult males, your population is all adult males
    - Typically impossible to study entire populations
- Sample
    - Subset of the population
- Survey
    - Posing questions to the sample to learn about the population
- Representative
    - Samples reflect the mix of the entire population
- Bias
    - Systemic error in selecting the sample
    - E.g. ignoring a group of people or data
- Properties of sampling
    - The best way to get a representative sample is to pick the members of the population at random
    - The sample should be as large as possible, but larger populations do not require larger samples
- Estimating parameters
    - Parameters
        - A characteristic of the population
        - Can be referred to as the "true" mean, sd, etc
        - Typically use greek letters like µ
    - Statistic
        - An observed characteristic of a sample
        - Doesn't use greek letters, instead symbols like $\bar{y}$ to show it is a characteristic of a sample 
        - Statistics, like the mean, can vary from sample to sample
            - $\bar{X_{1}}, \bar{X_{2}}...,\bar{X_{n}}$
    - Estimate
        - Using a statistic to approximate a parameter
    - Standard error
        - Variation from sample to sample which is the error of the estimate
        - Used instead of standard deviation for samples

### Notation for Statistics and Parameters

| Name               | Sample Statistic | Population Parameter |
|--------------------|------------------|----------------------|
| Mean               | $\bar{y}$        | μ                    |
| Standard deviation | s                | σ                    |
| Correlation        | r                | ρ                    |
| Slope of line      | b                | β                    |
| Proportion         | $\hat{p}$        | p                    |


## Sampling Distribution of the Mean
- Finding the means of multiple samples helps normalize individual distributions
    - However, central limit theorem tells us this is unnecessary
    - With a large enough sample size, we can just take the mean ($\bar{X}$) of a single sample
- $\bar{X}$ tends to be normally distributed
    - Low variation from sample to sample

## Central Limit Theorem (CLT)
- Regardless of the shape of the population, if the sample size is large enough the shape of the sample means tends to be normally distributed
    - Even if population is highly skewed
- Sampling distribution becomes normal as n increases even if the population is not normal

### CLT Properties
- Central tendency:
    - $µ_{\bar{X}} = µ$
- Standard error:
    - $σ_{\bar{X}} = \frac{σ}{\sqrt{n}}$
- Notation
    - $\bar{X} \sim N \left(µ, \frac{σ^{2}}{n} \right) $
- Sample size condition
    - A normal model provides an accurate approximation to the sampling distribution of $\bar{X}$ if the sample size *n* is **both**:
        - Larger than 10 times the squared skewness **and**
        - Larger than 10 times the absolute value of the kurtosis
    - $n > 10K^{2}_{3}$  **AND**  $n > 10|K_{4}|$


### Example: Highly Skewed Population
*A life insurance policy that pays \\$100k if a perforson dies in the next five years and collects \\$1k as payment at the beginning of every year.*
- Notice how skewed data is with n=30, nowhere near normal 

```r
# Sample size, the key to CLT
# In this example it refers to the size of the pool of insured customers
# Adjust this number to explore CLT
n = 30
#n = 100
#n = 1000

# NSIM refers to a set of potenetial random future outcomes of the pool of n customers
# The number of simulations we will use later
nsim = 10000

# The probability model of the insurance company
# y refers to thousands, p is probability
# The first item can be read as ".02% probability of losing $99(thousand)"
# Last item is a 99% chance of gaining $5(thousand) so this is highly skewed
y = c(-99, -98, -97, -96, -95, 5)
p = c(.002, .002, .002, .002, .002, .99)

# ntotal is the number of simulations with sample size n
ntotal = n * nsim

# Create the simulation
rv = sample(y, ntotal, p, replace = TRUE)

# Create a matrix with number of columns equal to sample size n
rvm = matrix(rv, ncol=n, byrow = TRUE)

# Get the means of each row (each row is a sample)
average = rowMeans(rvm)

# Create histogram
# Use main to create title with variable n
hist(average, freq=FALSE, main = paste('Distribution of Average of', n, 'policies'))
```

Let's examine this again with a higher n:

Let's examine this again with a higher n:
- The distribution now looks much more normal
- We can further examine with qqnorm and qqline to see normality


```r
# Adjust this number to explore CLT
#n = 30
#n = 100
n = 1000

# The number of simulations we will use later
nsim = 10000

# y refers to thousands, p is probability
y = c(-99, -98, -97, -96, -95, 5)
p = c(.002, .002, .002, .002, .002, .99)

# ntotal is the number of simulations with sample size n
ntotal = n * nsim

# Create the simulation
rv = sample(y, ntotal, p, replace = TRUE)

# Create a matrix with number of columns equal to sample size n
rvm = matrix(rv, ncol=n, byrow = TRUE)

# Get the means of each row (each row is a sample)
average = rowMeans(rvm)

hist(average, freq=FALSE, main = paste('Distribution of Average of', n, 'policies'))

qqnorm(average)
qqline(average)
```

#### Find Sample Statistics
- The true mean gives us the expected amount of profit (\\$3,980)
- The standard error would be $\pm$ \\$321

```r
# Calculate true mean given from probability model
mu = sum(y*p)
print(paste('The true mean is: ', mu))

# Find variance to use in stderror 
var = sum((y-mu)^2 * p)

# Find standard error with n = 1000
stderror = sqrt(var)/sqrt(1000)

print(paste('The standard error is: ', stderror))
```

## Sampling Distribution for Proportions
- Used for categorical data, rather than numeric
- E.g. "Proportion of people who are vegetarians"
    - Categorical because answers are yes or no

### Properties
- Mean of $\hat{p}$
    - $µ_{\hat{p}} = p$
    - Useful is population *p* is known
        - If *p* isn't known we can use a small sample
- Standard error
    - $σ_{\hat{p}} = \sqrt{\frac{p(1-p)}{n}}$

### Example: Probability of Proportion
*The true proportion of Texas students who drive a truck is p = 0.4, what is the probability that a sample size of 200 yields a sample proportion between 0.4 and 0.5*
- Find µ and $\hat{p}$ to find probability
- Use pnorm(0.45, mean, se) - pnorm(0.4, mean, se) to get the probability between the values
    - 0.43

```r
p = 0.4

n = 200

# P(0.4 < phat < 0.45)

# Remember µ of p-hat is equal to population p
mean = 0.4
se = sqrt((p*(1-p))/n)

print(paste('Mean: ', mean))
print(paste('Standard error: ', se))

prob = pnorm(0.45, mean, se) - pnorm(0.4, mean, se)

print(paste('The probability of proportions between 0.4 and 0.45 is ', prob))
```
