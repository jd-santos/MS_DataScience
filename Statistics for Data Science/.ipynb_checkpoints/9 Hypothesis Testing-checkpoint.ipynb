# Hypothesis Testing

## Hypothesis Overview
- A hypothesis is a testable statement about a population parameter 
- Population mean notation
    - The mean of x is more than 150
        - µ $\geq$ 150
    - The mean of x = y
        - $µ_{1} = µ_{2}$
- Population proportion notation
    - The proportion of adults in Lubbock with a smart watch is 18%
        - p = 0.18
    - The porportion of adults in Austin with a smart watch is at least more than adults in Lubbock
        - $p_{1} \geq p_{2}$
        
- The Null Hypothesis, $H_{0}$
    - Represents the current belief in a situation
    - Begin with the assumption that null hypothesis is true
        - May be accepted or rejected
    - Use =, $\geq$, or $\leq$
- The Alternative Hypothesis, $H_{a}$
    - Opposite of the null hypothesis
    - Never contains =, $\geq$, or $\leq$
        - Use !=, >, or <

## Hypothesis Test for Population Proportion
- Often expressed as percentage 
- Convert p to z-scores
    - Find if z- score is out of range

```r
display_png(file="Images/08.3_Proportion.png") 
```

In this example, we are trying to determine if a spam filter is effective at reducing spam to under 20%
- This is a *one-sided test*
- $H_{0} = p \geq 0.2 $
- $H_{a} = p < 0.2$
- The null distribution has the parameters:
    - $µ\hat{p} = p_0$
    - $SE\hat{p} = \sqrt{\frac{p_{0}(1-p_{0})}{n}}$
- Find $Z_{stat}$ with:
    - $Z_{stat} = \frac{\hat{p} - p_{0}}{SE}$

A review of 100 emails showed 12% spam.  Is this just chance?

```r
# Null hypothesis
p0 = 0.2

# Sampling information
n = 100

# Proportion (p-hat)
phat = 0.12

# Checklist: The sample size is large enough to be  
#... approximated by normal distribution if: 
n * phat > 10
n * (1-phat) > 10

# Null distribution of p-hat has the following parameters:
# µ:
muPhat = p0
# Standard error of p-hat:
sigmaPhat = sqrt(p0*(1-p0)/n)

# Find the test statistic (z-score)
# z is always equal to (xµ/sigma)
Zstat = (phat - p0) / sqrt(p0 * (1-p0)/n)
print(paste('Z-score: ',Zstat))
```

Now we can check if the Z-stat falls in the critical region:

```r
alpha = 0.05

print(paste('Z-score: ',Zstat))
# Since this is on the left tail, we need to compare with -qnorm
critregion = -qnorm(1-alpha)
print(paste('Critical region: ',critregion))
Zstat < critregion
```

The Z-score is less than the leftmost critical region or 2 standard errors below $p_{0}$. Since z < -1.65 we **reject** the null hypothesis and the result is statistically significant.

We can examine this graphically:

```r
# Make sequence of numbers for x-axis:
z = seq(-4, 4, 0.01)

# Plot z, normal dist of z
plot(z, dnorm(z), main = "H0 distribution", type = 'l', yaxs = 'i')

# Critical value line:
abline(v = -qnorm(1-alpha))

# Z-stat
abline( v = Zstat, col = 'red')
```

#### P-Value
- In this context the p-Value is the smallest α level at which $H_{0}$ can be rejected

Finding p-value
- The probability that z is less than Zstat
- Here P(Z < $z_{stat}$) = P(Z < -2) = 0.0227
    - Observed level of significance, this small value means $H_{0}$ is not plausible

```r
# Find P-value with pnorm
print(paste('P-value: ', pnorm(Zstat, 0, 1) ))
```

## Hypothesis Tests for Population Mean
- Similar to population proportion with numerical rather than categorical data
- $T_{stat} = \frac{\bar{X}-µ_{0}}{SE}$
- SE = $\frac{s}{\sqrt{n}}$

```r
display_png(file="Images/08.5_Mean_Test.png") 
```

Example: To approve a park expansion the average rent in San Francisco needs to be \\$1500, are rents high enough to justify expansion?
- n = 115
- $\bar{X}$ = \\$1657
- s = \\$580

Solving:
- $H_{0}: µ \leq 1500$
- $H_{a}: µ > 1500$
- The null distribution of $\bar{X}$ can be approximated by the t distirubtion and has the following parameters:
    - $µ_{\bar{X}} = µ_{0}$
    - $SE_{\bar{X}} = \frac{s}{\sqrt{n}}$
- Find $T_{stat}$
    - $\frac{\bar{X} - µ_{0}}{\frac{s}{\sqrt{n}}}$
- Quantile 
    - `qt(1-α, df)`

```r
# Null hypothesis
mu0 = 1500

# Sample information
n = 115
xbar = 1657
s = 581

# Null distribution parameters
muXbar = mu0
sigmaXbar = s/sqrt(n)

# Find test statistic
Tstat = (xbar - mu0)/(s/sqrt(n))
print(paste('Tstat: ', Tstat ))

# Check significance
alpha = 0.05
df = n-1

# Use > because of alt hypothesis, looking at right tail
print(paste('Critical value: ', qt(1-alpha, df) ))
print(paste('Is Tstat > critical value? ', Tstat > qt(1-alpha, df) ))
```

We can reject the null hypothesis.

Finding p-value
- In this case area under the curve on the right side
- Use **pt(t, df)**
- For P( T > Tstat )
    - $1 - P(T\leq T_{stat})$
- Calculate by hand:
    - $ t = \frac{1,657 - 1,500}{\frac{581}{\sqrt{115}}}$
    - t = 2.898 with 114 df
        - Gives p-value of 0.0023

```r
print(paste('Probability that T > Tstat: ', 1 - pt(Tstat, df) ))
```

## Hypothesis Tests for Two Population Proportions

```r
display_png(file="Images/08.7_Hyptest_Two_Prop.png") 
```

**Two-Sample z-statistic:**

- $z = \frac{(\hat{p_{1}} - \hat{p_{2}}) - D_{0}}{se(\hat{p_{1}}-\hat{p_{2}})} $

- Using the estimated standard error of the difference between the sample proportions
    - $\sqrt{\frac{\hat{p_{1}}(1 - \hat{p_{1}})}{n_{1}} + \frac{\hat{p_{2}}(1 - \hat{p_{2}})}{n_{2}}}$
- As before, SRS and sample size checklists apply

```r
display_png(file="Images/08.7_TwoZ_Summary.png") 
```

Example: Let $p_{A}$ denote population proportion on Atkins diet who renew membership and $p_{C}$ proportion on conventional diet who renew membership.

The difference $p_{Α} - p_{C}$ measures the extra proportion who renew if on the Atkins diet. To be profitable this difference must be more than 4%
- $D_{0}$ will be 5 pounds
- $H_{0}: p_{Α} - p_{C} \leq 0.04$
- $H_{A}: p_{Α} - p_{C} > 0.04$

- Null distribution parameters:
    - $µ_{\hat{p_{1}}} - µ_{\hat{p_{2}}} = D_{0}$
    - $se({\hat{p_{1}}} - {\hat{p_{2}}}) = \sqrt{\frac{\hat{p_{1}}(1 - \hat{p_{1}})}{n_{1}} + \frac{\hat{p_{2}}(1 - \hat{p_{2}})}{n_{2}}}$

```r
# Null hypothesis
null.diff = 0.04

# Sampling information (given by example)
# 72% of 150 Atkins users renewed membership
n1 = 150
phat1 = 0.72

# 60% of 220 regular dieters renewed membership
n2 = 220
phat2 = 0.60

# Get null distribution parameters µ and se
muPhat.diff = 0.04
sigmaPhat.diff = sqrt(phat1*(1-phat1)/n1 + phat2*(1-phat2)/n2)

# Find Zstat
Zstat = (phat1 - phat2 - null.diff)/(sqrt(phat1*(1-phat1)/n1 + phat2*(1-phat2)/n2))
print(paste('The Zstat is: ', Zstat))

# Check if Zstat falls in critical region
# One-sided test
alpha = 0.05
critregion = qnorm(1-alpha)
print(paste('Zstat is outside the critical region: ', Zstat > critregion))
```

**We cannot reject the null hypothesis.**

Looking at this graphically:
- Use positive qnorm for right tail
- Z-stat is just inside the acceptance zone

```r
# Make sequence of numbers for x-axis:
z = seq(-4, 4, 0.01)

# Plot z, normal dist of z
plot(z, dnorm(z), main = "H0 distribution", type = 'l', yaxs = 'i')

# Critical value line:
# Positive qnorm for right tail
abline(v = qnorm(1-alpha))

# Z-stat
abline( v = Zstat, col = 'red')
```

Notice P-value is a little over .05

```r
#Get P-value
print(paste('The p-value is: ',1-pnorm(Zstat)))
```

### Confidence Interval for Two Proportions
$(\hat{p_{1}} - \hat{p_{2}}) \pm Z_{α/2}se(\hat{p_{1}} - \hat{p_{2}})$

```r
display_png(file="Images/08.75_CI_Two_Prop.png") 
```

When the 95% confidence interval does not include 0, we say the two proportions are statistically significantly different from each other. 

The Atkins members renew at a statistically significant higher rate.

## Hypothesis Tests for Two Population Means

**Compute t-statistic**
- $t = \frac{(\bar{X_{1}} - \bar{X_{2}}) - D_{0}}{se(\bar{X_{1}}-\bar{X_{2}})} $
- Conditions
    - SRS
    - Similar variances
    - Sample size condition
        - n1 > 10|Kurtosis2|
        - n2 > 10|Kurtosis2|
- In R use qt(alpha, df)
- Find p-value with pt(Tstat, df)

```r
display_png(file="Images/08.8_2sample_mean_df.png") 
```

```r
display_png(file="Images/08.8_2sample_mean_summary.png") 
```

Example: Let $μ_{A}$ denote the mean weight loss for Atkins diet and $μ_{C}$ denote weight loss for conventional diet.
- $D_{0}$ will be 5 pounds
- $H_{0}: μ_{Α} - μ_{C} \leq 5$
- $H_{A}: μ_{Α} - μ_{C} > 5$

```r
# Null hypothesis
null.diff = 5

# Sampling information (given by example)
# 33 Atkins dieters lost an average of 15.41lbs with 14.37lb sd
n1 = 33
xbar1 = 15.42
s1 = 14.37

# Conventional dieters:
n2 = 30
xbar2 = 7
s2 = 12.36

# Get null distribution parameters µ and se with xbar diff
xbar.diff = xbar1 - xbar2
muxbar.diff = null.diff
sigmaXbar.diff = sqrt(s1^2/n1 + s2^2/n2) 

# Find Tstat
Tstat = (xbar.diff - null.diff)/(sqrt(s1^2/n1 + s2^2/n2))
print(paste('The Tstat is: ', Tstat))

# Write function to get df
df.t <- function(s1, s2, n1, n2){
    nom = (s1^2/n1 + s2^2/n2)^2
    denom = (s1^2/n1)^2/(n1-1) + (s2^2/n2)^2/(n2-1)
    nom/denom
}

df = df.t(s1, s2, n1, n2)
print(paste('There are ',df, ' degrees of freedom'))

# Check if Tstat falls in critical region
# One-sided test
alpha = 0.05
critregion = qt(1-alpha, df)
print(paste('Tstat is outside the critical region: ', Tstat > critregion))

```

We fail to reject the null hypothesis

Examine graphically and find P-value:

```r
# Make sequence of numbers for x-axis:
t = seq(-4, 4, 0.01)

# Plot t, normal dist of t
plot(t, dnorm(t), main = "H0 distribution", type = 'l', yaxs = 'i')

# Critical value line:
# Positive qnorm for right tail
abline(v = qt(1-alpha, df))

# Z-stat
abline( v = Tstat, col = 'red')

# P-value
print(paste('The p-value is: ',1-pt(Tstat, df)))
```

### Confidence Intervals of Two Population Means
$(\bar{X_{1}} - \bar{X_{2}}) \pm t_{α/2}se(\bar{X_{1}} - \bar{X_{2}})$

## Two-Sample Test for Paired Comparison
- Comparing means for dependent populations
- Given paired data, we begin analysis by forming the difference within each pair
    - $d_{i} = x_{i} - y_{i}$
- Confidence paired t-interval
    - $\bar{d} \pm t_{α/2; n-1}\frac{s_{d}}{\sqrt{n}}$ with n-1df
    - d = differences
    - Calculating CI allows us to hypothesis test

Example: Comparing two sales forces within a company. Determine 95% confidence t-interval to find difference between two.
- $μ_{D} = µ_{A} - µ_{B} $
- $H_{0}: µ_{D} = 0$
- $H_{0}: µ_{D} \neq 0$

```r
display_png(file="Images/08.9_Paired_Comparison.png") 
```

```r
# Null hypothesis
null.diff = 0

# Sampling information (given by example)
# 72% of 150 Atkins users renewed membership
n1 = 230
phat1 = 0.24

# 60% of 220 regular dieters renewed membership
n2 = 269
phat2 = 0.30

# Get null distribution parameters µ and se
muPhat.diff = 0
sigmaPhat.diff = sqrt(phat1*(1-phat1)/n1 + phat2*(1-phat2)/n2)

# Find Zstat
Zstat = (phat1 - phat2 - null.diff)/(sqrt(phat1*(1-phat1)/n1 + phat2*(1-phat2)/n2))
print(paste('The Zstat is: ', Zstat))

# Check if Zstat falls in critical region
# One-sided test
alpha = 0.1
critregion = qnorm(1-alpha)
print(paste('Zstat is outside the critical region: ', Zstat > critregion))

(phat1-phat2) - 1.645 * sigmaPhat.diff
(phat1-phat2) + 1.645 * sigmaPhat.diff

```

## Hypothesis Test for Indepedence ($X^{2}$)

### Overview
- Chi-Squared Test used to check for dependence between variables
- Hypotheses:
    - $H_{0}$: Variables are independent
    - $H_{a}$: Variables are not independent
- Test Statistic ($X^{2}$)
    - Measures the distance between the observed contingency table and a hypothetical contingency table
    - The hypothetical contingency table follows the null hypothesis
        - Assumes two random variables are completely independent

### Testing in R
- Import some simple income and purchase data
- Convert to contingency table
- Create a null hypothesis table with independent rows and columns

```r
data <- read.csv("http://tiny.cc/amazonStat")
head(data)

# Add new column to data with income levels
# Define the intervals based on the income column
# Begins with -(Inf)inity and ends with (Inf)inity
data$incomeLevel = c("L1:25orLess", "L2:26to50", "L3:51to75", "L4:76to100", "L5:100orMore")[
                findInterval(data$income, c(-Inf, 25000, 50000, 75000, 100000, Inf))]
head(data)

# H0: Income and product category are independent of each other
# Ha: Income and Product category are not independent of each other

# Number of Observations
n = length(data$income)
n

# Create contingency table
obs.table <- table(data$category, data$incomeLevel)
obs.table

# Mosaic plot shows us proportions of people who buy each item
mosaicplot(t(obs.table), col = 2:3)
```

Create a null hypothesis table:
- Remember definition of independence is the product marginal probabilities over n
    - We represent this here with rowsum * colsum / n
- We can confirm the table is perfectly null with the mosaic

```r
# Get row and column sums
rowsum = rowSums(obs.table)
colsum = colSums(obs.table)
exp.table <- obs.table

# Change the table cells to the rowsum*colsum/n
# rowsum*colsum = product of marginal probabilities
# This gives us perfectly independent table
for (i in 1:nrow(obs.table)) {
    for (j in 1:ncol(obs.table)) {
        exp.table[i,j] = rowsum[i] * colsum[j]/n
    }
}

# View as Mosaic
mosaicplot(t(exp.table), col = 2:3)
```

### Calculating $X^{2}$
- $X^{2} = \sum\frac{(observed - expected)^{2}}{expected}$
    - Expected is the hypothetical independent data
    - Summing all values of $\frac{(o-e)^{2}}{e}$ from every cell gives overall $X^{2}$
- Checklist
    - Ensure sample is based on SRS
    - Rows and columns must be mutually exclusive (no intersection) 
- Degrees of Freedom
    - df = (r-1)(c-1)
    - r and c: number of rows and columns

```r
library("IRdisplay")
display_png(file="Images/9_Chi_Dist.png") 
```

### Chi-Squared in R
- Use qchisq(alpha, df) to find critical value
- Compare $X^{2}$ to critical value
- Here we find $X^{2}$ is much larger than the critical value 

```r
# X-square statistic
x2stat = sum((obs.table-exp.table)^2/exp.table)
print(paste('Chi-squared is: ', x2stat))
alpha = 0.05

# df = (rows-1)(cols-1)
df = (nrow(obs.table)-1) * (ncol(obs.table)-1)

# Critical value
x2crit = qchisq(1-alpha, df)
print(paste('Critical value is: ',x2crit))

# Compare chi-square and critical value
x2result = x2stat > qchisq(1-alpha, df)
print(paste('Chi-squared is greater than critical value: ',x2result))
```

#### Graphing Result
- This shows the chi-square is far outside the critical value
- We can strongly reject the null hypothesis

```r
# Make sequence of numbers for graph
x = seq(0, 40, 0.01)

# Plot distribution
plot(x, dchisq(x, df), main = "H0 distribution", type = 'l', yaxs = 'i')

# Critical value
abline(v = qchisq(1-alpha, df))

# Chi-square (red)
abline(v = x2stat, col = 'red')
```

#### Finding P-value
- Uses pchisq(xstat, df)
- P-value = P(x2 > x2stat) = 1 - pchisq
- We see the p-value is extremely small, close to zero
- P-value < α allows us to reject null

```r
1-pchisq(x2stat, df)
```

#### Use chisq.test Function
- Easier method without doing above calculations

```r
test = chisq.test(obs.table)
test
```
