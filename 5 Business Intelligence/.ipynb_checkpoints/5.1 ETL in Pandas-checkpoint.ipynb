# 5.1 ETL in Pandas Overview

## What is Pandas?
- Python library that eases data related tasks in Python
- Effectively creates a 2D spreadsheet
- At the time of these notes, Pandas is under very active development. Check documentation frequently

## Data Structure
- Dataframe
    - The primary structure of Pandas
    - Two-dimensional
    - Can reference column or rows
    - Frequently abbreviated df 
- Series
    - One-dimensional array
    - Can be converted to a dataframe 
    - There is a difference between a 1-column dataframe and a series

## Reading and Writing

### Module and Variable Setup

- Import pandas, commonly aliased to pd
- Import io, this will be used to hold get requests in memory for reading
- Import os to access files

```python
import io
import os
import pandas as pd
import requests as r
```

Set some variables for file access
- `os.path.abspath("notebookname.ipynb"`
    - Stores the absolute path of your notebook as a variable "path"
- `os.path.join(os.path.dirname(path), "relative-file-path")`
    - Combines the absolute path already created with any relative path of the notebook you stored in the above variable 
- Both of these together allow us to save the full path of our data output file in variable file_out

```python
url = 'http://drd.ba.ttu.edu/isqs6339/ex/L2.1/'
path = os.path.abspath("5.1 ETL in Pandas.ipynb")
file_1 = 'test_data.csv'
file_2 = 'test_data_pipe.csv'
file_out = os.path.join(os.path.dirname(path), "DataOut/5.1-outfile.csv")
```

### Basic File Read from Local Disk

Use read_csv from pandas to read the csv into a dataframe called df:
- But I don't actually have this file so this code is lies

```python
# The csv is a lie
df = pd.read_csv(path + file_1)
df
```

### File Read from Web Request

We don't have the CSV, but our nice professor is kind enough to host it on his site.
- Use a get request and store the file in 'res'
- Confirm with status code

```python
res = r.get(url + file_1)
res.status_code
df = pd.read_csv(io.StringIO(res.text))
df
```

```python
res = r.get(url + file_2)
res.status_code
```

`io.StringIO()`
- pd.read_csv is looking for a local file, but we don't have that
- io.StringIO() will write a string to an in-memory buffer
    - This emulates having a file for read_csv to read
    - We pass in res.text to get the text of res
- Use Pandas read_csv to store our requested csv into 'df' dataframe

```python
df = pd.read_csv(io.StringIO(res.text))
df
```

### File Read from Local with Delimiter Change
- When we ran the previous example, all of our data got shoved into a single column
- read_csv defaults to comma delimiter
- Add extra parameter to change this to | 
- No need to re-run get request because it was stored in 'res'

```python
df = pd.read_csv(io.StringIO(res.text), delimiter='|')
df
```

### File Write Output
- Check the [Pandas CSV Export Documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html)
- Known issue when translating between OS's
- `df.to_csv`
    - Write out to csv

```python
df.to_csv(file_out)
```

An extra index column is added to the beginning of a file by default
    - Remove the row index with `index=False`parameter
    - Change the delimiter to pipes with `sep` parameter

```python
df.to_csv(file_out, sep='|', index=False)
```
