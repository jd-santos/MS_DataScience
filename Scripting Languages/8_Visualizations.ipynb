### Random Time-Series Plot
- Numpy gives us modules to create random numbers
- **numpy.random.rand(d0,d1,...dn)**
    - Creates random values from 0-1 in a given shape
    - Parameters are optional
        - If given they specify the shape of the random numbers generated
        - Ex. (100) gives you 100 random numbers in one dimension
        - Ex. (3,2) gives you random numbers in 3 rows and 2 columns
- **numpy.random.randn(d0,d1,...dn)**
    - Return samples from standard normal distribution
    - Randn, n for normal :) 
    - Same parameter types as rand

```python
# Rand variables in 2x3
np.random.rand(3,2)
```

```python
# Randn 10 variables in one dimension
np.random.randn(10)
```

Here we will create a random normal distribution with 1096 variables, use a date range as the index, and calculate the cumulative summary:

```python
# Create seed variable to easily seed random number sequence
# Same seeds give same random variables for reproducible code
seedval = 111111

# Use numpy random to create random series
# Creates random normal dist of 1096 variables
# Uses date range as index
np.random.seed(seedval)
s = pd.Series(np.random.randn(1096),
             index = pd.date_range ('2015-01-01',
                                   '2017-12-31'))

walk_ts = s.cumsum()

# Plot the cumulative summary
walk_ts.plot()
```

A good practice is to make these into Dataframes so it can support multiple columns:

```python
df_walk = pd.DataFrame(walk_ts)
df_walk.plot()
```

### Multi-Column Plot

Here we will modify the code slightly to generate two random, with each walk as a column of the dataframe:
  - Notice this time we add a second parameter to randn to create a second dimension (1096,2)
  - Use list('AB') as column value to label the two columns 'A' and 'B'

```python
np.random.seed(seedval)
df = pd.DataFrame(np.random.randn(1096,2),
            # Use the index from walk_ts
             index = walk_ts.index,
                  columns = list('AB'))

# Set df_walk to the cumulative summary of 'df' above
df_walk = df.cumsum()

# Get head
df_walk.head()

# Plot
df_walk.plot()
```

#### Change X-axis from Dates to Values
- Create a third column of values with arange
    -Use len([dataframe]) to get range of values created above
- numpy.arange([start],[stop],[step],[dtype]
    - Return list of evenly spaced values
    - Include start value but exclude stop
    - Datatype defaults to none, infer from other input
- Use that third column as the X-axis

```python
# Copy the Dataframe to new df
df2_walk = df_walk.copy()

# Add column C which is list from 0 to the length of df (1096)
# Use same index 
df2_walk['C'] = pd.Series(np.arange(0, len(df2_walk)),
                         index = df2_walk.index)

# Set the x axis to C
# This makes the x-axis a label from 0 to 1096
# Set y-axis to A and B
df2_walk.plot(x = 'C', y = ['A', 'B'])
```

### Matplotlib Overview
- Plotting library designed to be similar to Matlab
- import matplotlib.pyplot 
    - Typically aliased to plt
- plt.plot()
    - Arguments: (x, y, formatting args)
    - See [matplotlib.pyplot.plot documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)

```python
import matplotlib.pyplot as plt

# Plot with lists of four x and y values respectively
plt.plot([1,2,3,4], [5,6,7,8])
```

```python
x = range(0,100)
y = [v*v for v in x]

# Formatted with red o's ('ro')
plt.plot(x,y,'ro')
```

### Heatmap
- Values within a matrix are represented by colors
- You need to create a correlation matrix
- Uses matplotlib
    **.imshow(heatmap_data, cmap = 'hot', interpolation = 'none')**
- Add colorbar with **plt.colorbar()**
- Add index labels with **plt.xticks/yticks**

```python
s = pd.Series([0.0,0.1,0.2,0.3,0.4],
             ['V','W','X','Y','g'])

heatmap_data = pd.DataFrame({'A': s + 0.0,
                            'B': s + 0.1,
                             'C': s + 0.2,
                             'D': s + 0.3,
                             'E': s + 0.4,
                             'F': s + 0.5,
                             'G': s + 0.6
                            })
heatmap_data

# Create heatmap with plt.imshow function with these arguments:
plt.imshow(heatmap_data, cmap = 'hot', interpolation = 'none')

# Add Colorbar
plt.colorbar()

# Add x/y-axis indexes
# For x grab column names
plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns)
# For y grab the row names (default of heatmap_data is row)
plt.yticks(range(len(heatmap_data)), heatmap_data.index)
```
